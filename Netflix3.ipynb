```python
# Writing function to convert all the values of date_added column into datetime

def convert_date(date_str):
    # Check if the input is a string
    if isinstance(date_str, str):
        # Remove leading spaces
        date_str = date_str.lstrip()
        try:
            return datetime.strptime(date_str, "%B %d, %Y")
        except ValueError:
            # Print a warning and return None for invalid date strings
            print(f"Warning: Could not convert '{date_str}' to datetime.")
            return None
    else:
        # Handle non-string values (e.g., NaN)
        print(f"Warning: Input is not a string: '{date_str}'")
        return None

# Apply the function to the 'date_added' column
df['date_added'] = df['date_added'].apply(convert_date)
```

```python
# Creating another dataframe that contains all rows that have null value in data_added column
df_1 = df[df['date_added'].isnull()]
```

```python
import pandas as pd

def top3_genres_by_country(df):

    country_genre_counts = df.groupby(['country', 'genres'])['genres'].count().reset_index(name='count')

    country_genre_counts = country_genre_counts.sort_values(['country', 'count'], ascending=[True, False])

    top3_genres = country_genre_counts.groupby('country').head(3)

    return top3_genres

# Assuming your DataFrame is named 'df' and contains 'country' and 'genres' columns
top3_df = top3_genres_by_country(df)
print(top3_df)
```

```python
import plotly.express as px

# Assuming you have the 'top3_df' DataFrame from the previous code
fig = px.bar(top3_df,
             x="country",
             y="count",
             color="genres",
             title="Top 3 Genres by Country",
             labels={"country": "Country", "count": "Number of Titles", "genres": "Genre"},
             barmode='stack')  # Grouped bar chart for better comparison

fig.update_layout(xaxis_tickangle=-45)  # Rotate x-axis labels for readability

fig.show()
```
